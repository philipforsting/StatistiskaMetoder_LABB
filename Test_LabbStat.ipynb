```python
import pandas as pd
import numpy as np
import LabbStat

import matplotlib.pyplot as plt # MÃ…STE BORT

housing_df = pd.read_csv('housing.csv')
X = housing_df['median_income']
y = housing_df['median_house_value']

model = LabbStat.LinjearRegression()
model.fit(X, y)


```

```python
model.beta_hat
```

```python
model.variance()
```

```python
model.standard_deviation()
```

```python
model.root_mean_squared_error()
```

```python
plot = True
if plot:
    plt.scatter(model.X[:, 1], model.y)
    x = np.linspace(min(X), max(X))
    y_hat = model.beta_hat[0] + model.beta_hat[1]*x
    plt.plot(x, y_hat, c="red")
```
