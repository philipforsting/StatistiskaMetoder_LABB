```python
import pandas as pd
import numpy as np
import LabbStat

import matplotlib.pyplot as plt # MÃ…STE BORT

housing_df = pd.read_csv('housing.csv')
X = housing_df['median_income']
y = housing_df['median_house_value']

model = LabbStat.LinjearRegression()
b = model.fit(X, y)
variance = b.variance()
standard_deviation = b.standard_deviation()
standard_deviation

```
